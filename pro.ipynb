```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dfd['Gender']=le.fit_transform(dfd['Gender'])
dfd['Family_Heart_Disease']=le.fit_transform(dfd['Family_Heart_Disease'])
dfd['Smoking']=le.fit_transform(dfd['Smoking'])
```

```python
x=dfd.drop(['Heart_Risk'],axis=1).values
y=dfd['Heart_Risk'].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_pred,y_test)
ac
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred)
print(cr)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
lab=[0,1]
cmd=ConfusionMatrixDisplay(cm,display_labels=lab)
cmd.plot()
```

```python
import pickle

# Save model
with open("svm_heart_model.pkl", "wb") as f:
    pickle.dump(model, f)

# Save scaler
with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)

# Save column order
with open("columns.pkl", "wb") as f:
    pickle.dump(dfd.drop("Heart_Risk", axis=1).columns.tolist(), f)




```
